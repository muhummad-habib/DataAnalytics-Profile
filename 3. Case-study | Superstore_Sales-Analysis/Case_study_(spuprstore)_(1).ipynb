# Q1: What is the overall Sales trend?

```python
data.head(1)
```

```python
data['order_date'].min()
```

```python
data['order_date'].max()
```

```python
# Replace comma with empty space for sale attribute
data['sales']=data['sales'].str.replace(',','')
```

```python
# Change object data tipe into ineger using astype(int)
data['sales']=data['sales'].astype(int)
```

```python
data.info()
```

```python
# Convert order_date time object into real date & time using pandas
data['order_date']=pd.to_datetime(data['order_date'])
```

```python
# Getting months year from order_date
data['month_year']=data['order_date'].apply(lambda x:x.strftime('%Y-%m'))
```

```python
data['month_year']
```

```python
# Grouping month_year by sale
data_trend=data.groupby('month_year').sum()['sales'].reset_index()
```

```python
data_trend
```

```python
plt.figure(figsize=(15,6))
plt.plot(data_trend['month_year'],data_trend['sales'],color='#b80045')
plt.xticks(rotation='vertical',size=8)
plt.show
```
